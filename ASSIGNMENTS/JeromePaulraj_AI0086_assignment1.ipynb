```python
import numpy as np
import pandas as pd
from sklearn import tree,model_selection,metrics,ensemble,linear_model,preprocessing,naive_bayes,neighbors
```

```python
# data
data = pd.read_csv('train.csv')
data = data.iloc[:,2:] # Removing the unamed column, ID and the categorical column from the dataset
data
```

```python
X = data.iloc[:,:-1]# Removing the class column from the train data set
Y = data.iloc[:,-1]# Storing the classes

X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y,train_size=0.60, random_state=2)

# Normalizing the dataset to 1 and 0 (Using Min-Max)

x = X_train.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X_train = pd.DataFrame(x_scaled)

x = X_test.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X_test = pd.DataFrame(x_scaled)
X_train
```

```python
# DECISION TREE

decision_tree = tree.DecisionTreeClassifier(random_state=2)
decision_tree.fit(X_train,Y_train)
cv_dt = model_selection.cross_val_score(decision_tree, X_train, Y_train, cv=10, scoring='f1')
predicted_dt = decision_tree.predict(X_test)
score = metrics.f1_score(Y_test,predicted_dt)
print("F1 Score for decision tree: %0.2f (+/- %0.2f)" % (cv_dt.mean(), cv_dt.std() * 2))
print("F1 score: {0}".format(score))
```

```python
# Random Forest

random_forest = ensemble.RandomForestClassifier(random_state=2)
random_forest.fit(X_train,Y_train)
cv_rf = model_selection.cross_val_score(random_forest, X_train, Y_train, scoring='f1',cv=10)
predicted_rf = random_forest.predict(X_test)
score = metrics.f1_score(Y_test,predicted_rf)
print("F1 Score for random forest: %0.2f (+/- %0.2f)" % (cv_rf.mean(), cv_rf.std() * 2))
print("F1 score: {0}".format(score))
```

```python
# LOGISTIC REGRESSION

logistic_reg = linear_model.LogisticRegression(random_state=2)
logistic_reg.fit(X_train,Y_train)
cv_lg = model_selection.cross_val_score(logistic_reg, X_train, Y_train, scoring='f1',cv=10)
predicted_lg = logistic_reg.predict(X_test)
score = metrics.f1_score(Y_test,predicted_lg)
print("F1 Score for logistic regression: %0.2f (+/- %0.2f)" % (cv_lg.mean(), cv_lg.std() * 2))
print("F1 score: {0}".format(score))
```

```python
# Naive bayes
nb = naive_bayes.GaussianNB()
nb.fit(X_train,Y_train)
cv_nb = model_selection.cross_val_score(nb, X_train, Y_train, scoring='f1',cv=10)
predicted_nb = nb.predict(X_test)
score = metrics.f1_score(Y_test,predicted_nb)
print("F1 Score for Naive Bayes using cross validation: %0.2f (+/- %0.2f)" % (cv_nb.mean(), cv_nb.std() * 2))
print("F1 score: {0}".format(score))
```

```python
# KNN
knn = neighbors.KNeighborsClassifier()
knn.fit(X_train,Y_train)
cv_knn = model_selection.cross_val_score(knn, X_train, Y_train, scoring='f1',cv=10)
predicted_knn = knn.predict(X_test)
score = metrics.f1_score(Y_test,predicted_knn)
print("Score for KNN with k=5 using cross validation: %0.2f (+/- %0.2f)" % (cv_knn.mean(), cv_knn.std() * 2))
print("F1 score: {0}".format(score))
```
